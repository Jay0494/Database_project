```python
import tkinter as tk
from tkinter import messagebox
import mysql.connector
from datetime import datetime

# Function to submit data to MySQL
def submit_data():
    buyer_id = entry_buyer_id.get()
    item = entry_item.get()
    item_category = entry_item_category.get()
    service_provider = entry_service_provider.get()
    price = entry_price.get()
    quantity = entry_quantity.get()
    date = entry_date.get()

    # Input validation
    if not all([buyer_id, item, item_category, service_provider, price, quantity, date]):
        messagebox.showwarning("Input Error", "Please fill all fields")
        return

    try:
        # Connecting to the MySQL database
        conn = mysql.connector.connect(
            host="127.0.0.1",
            user="jaytech",
            password="Shaun@eliana123",
            database="elyon_database"
        )
        cursor = conn.cursor()

        # Inserting data into the purchase table
        sql = """INSERT INTO purchase 
                 (Buyer_Id, Item, Item_category, Service_provider, Price, Quantity, Date) 
                 VALUES (%s, %s, %s, %s, %s, %s, %s)"""
        values = (buyer_id, item, item_category, service_provider, price, quantity, date)
        cursor.execute(sql, values)
        conn.commit()

        # Show success message and clear fields
        messagebox.showinfo("Success", "Data submitted successfully")
        clear_fields()

    except mysql.connector.Error as err:
        messagebox.showerror("Database Error", f"Error: {err}")

    finally:
        if conn.is_connected():
            cursor.close()
            conn.close()

# Function to clear input fields
def clear_fields():
    entry_buyer_id.delete(0, tk.END)
    entry_item.delete(0, tk.END)
    entry_item_category.delete(0, tk.END)
    entry_service_provider.delete(0, tk.END)
    entry_price.delete(0, tk.END)
    entry_quantity.delete(0, tk.END)
    entry_date.delete(0, tk.END)

# Creating the GUI window
root = tk.Tk()
root.title("Purchase Data Entry Form")

# Labels and entry fields
labels = [
    "Buyer Id:", "Item:", "Item Category:",
    "Service Provider:", "Price:", "Quantity:", "Date (YYYY-MM-DD):"
]
entries = []

# Creating input fields based on labels
for i, label in enumerate(labels):
    tk.Label(root, text=label).grid(row=i, column=0, padx=10, pady=5)
    entry = tk.Entry(root)
    entry.grid(row=i, column=1, padx=10, pady=5)
    entries.append(entry)

# Assigning entries to variables
entry_buyer_id, entry_item, entry_item_category, \
entry_service_provider, entry_price, entry_quantity, entry_date = entries

# Submit button
submit_btn = tk.Button(root, text="Submit", command=submit_data)
submit_btn.grid(row=len(labels), column=0, columnspan=2, pady=20)

# Run the GUI application
root.mainloop()
```
